### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here

median = cars['mpg'].median()
above_median_cars = cars[cars['mpg'] > median]
above_median_cars = above_median_cars.copy()
above_median_cars['hp/weight'] = above_median_cars['horsepower'] / above_median_cars['weight']
best_car = above_median_cars.loc[above_median_cars['hp/weight'].idxmax()]
print(best_car)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
plt.figure(figsize=(12, 6))
sns.lineplot(data= cars, x='model_year', y='mpg', hue='origin', marker='o')
plt.title('Average MPG Over Years by Origin')
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
# Reset index to default integer index
cars_reset = cars.reset_index()

plt.figure(figsize=(12, 8))
sns.scatterplot(data=cars_reset, 
                x='horsepower', 
                y='weight', 
                hue='origin', 
                size='mpg',
                hue_order=['japan', 'europe', 'usa'],
                sizes=(50, 200))

plt.title('Car Weight vs Horsepower by Origin and MPG', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Weight', fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
